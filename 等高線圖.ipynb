```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def f(x,y):
    return(1-x/2+x**5+y**3)*np.exp(-x**2 -y**2)
n=500
x=np.linspace(-3,3,n)
y=np.linspace(-3,3,n)
X,Y=np.meshgrid(x,y)
plt.contourf(X,Y,f(X,Y),8,alpha=0.75,cmap='viridis')
C=plt.contour(X,Y,f(X,Y),8,colors='black')
plt.clabel(C,inline=True,fontsize=10)
plt.xticks(())
plt.yticks(())
```
