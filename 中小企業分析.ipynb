```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("https://www.moeasmea.gov.tw/files/559/70C1CE33-8E24-4959-A0FA-5606CCA05478")
```

```python
df1=df.drop(df[df.行業別=="總計"].index)
```

```python
df1
```

### 中小企業佔比

```python
df1["中小企業佔比"]=df.中小企業/df.全部
```

```python
df1
```

### 排序(由高至低)

```python
df1=df1.sort_values(by=["中小企業佔比"],ascending=False)
```
