```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=np.linspace(0,10,1000)
y=np.tan(x)
z=1/((np.cos(x))**2)
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,label="$tan(x)$",color="blue",linewidth=2)
```

```python
plt.plot(x,z,"r--",label="$1/(cos(x)**2)")
```

```python
x=np.linspace(0,10,1000)
y=np.tan(x)
z=1/((np.cos(x))**2)
plt.figure(figsize=(8,4))
plt.plot(x,y,label="$tan(x)$",color="blue",linewidth=2)
plt.plot(x,z,"r--",label="$1/(cos(x)**2)")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(0,100)
plt.legend()
plt.show()
```
