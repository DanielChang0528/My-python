```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Tesla=[1456,1679,2130]
Porsche=[1567,1143,1654]
Ferrari=[1739,2048,1573]
```

```python
seq=[2017,2018,2019]
```

```python
plt.xticks(seq)
plt.plot(seq,Tesla,'-*',label='Tesla')
plt.plot(seq,Porsche,'-o',label='Porsche')
plt.plot(seq,Ferrari,'-,',label='Ferrari')
plt.legend(loc='best')
plt.title("Sales Report",fontsize=20)
plt.xlabel("Year",fontsize=10)
plt.ylabel("Number of Sales",fontsize=10)
plt.tick_params(axis='both',labelsize=10,color='blue')
plt.show()
```
