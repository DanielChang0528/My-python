```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sympy as sp
```

```python
left=-2*np.pi
right=4
x=np.linspace(left,right,100)
y=np.sin(2*x)
plt.plot(x,y)
plt.fill_between(x,0,y,color='blue',alpha=0.1)
plt.show()
x=sp.symbols('x')
sp.integrate(sp.sin(2*x),(x,-2*np.pi,4))
```
