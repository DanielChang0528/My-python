### 歷史系學測一階篩選

```python
df=pd.read_csv("http://bit.ly/gradescsv")
```

### 四科總級分

```python
df["總級分"]=df.國文+df.英文+df.數學+df.社會
```

### 刪去自然

```python
df.drop("自然",axis=1)
```

### 由總級分排序高低

```python
df_sorted=df.sort_values(by=["總級分"],ascending=False)
```

### 篩選出總級分>=49

```python
df_sorted_1=df_sorted[df_sorted.總級分>=49]
```

### 篩選出國文+英文>=27

```python
df_sorted_2=df_sorted_1[(df_sorted_1.國文)+(df_sorted_1.英文)>=27]
```
